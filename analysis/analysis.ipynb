```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_dqn_rewards = pd.read_csv("DQN_1_reward.csv", names=["Wall_time", "Step", "Value"], skiprows=1)

df_dqn_rewards["Wall_time"] = pd.to_numeric(df_dqn_rewards["Wall_time"])
df_dqn_rewards["Time_in_hours"] = (df_dqn_rewards["Wall_time"] - df_dqn_rewards["Wall_time"].iloc[0]) / 3600
df_dqn_rewards = df_dqn_rewards.loc[df_dqn_rewards["Time_in_hours"] <= 2.53]

display(df_dqn_rewards)
```

```python
df_a2c_rewards = pd.read_csv("A2C_1_reward.csv", names=["Wall_time", "Step", "Value"], skiprows=1)

df_a2c_rewards["Wall_time"] = pd.to_numeric(df_a2c_rewards["Wall_time"])
df_a2c_rewards["Time_in_hours"] = (df_a2c_rewards["Wall_time"] - df_a2c_rewards["Wall_time"].iloc[0]) / 3600

display(df_a2c_rewards)
```

```python
plt.style.use('default')


plt.figure(figsize=(8, 5))  # Width: 12 inches, Height: 8 inches
plt.plot(df_dqn_rewards["Time_in_hours"], df_dqn_rewards["Value"], label="DQN", marker="o")
plt.plot(df_a2c_rewards["Time_in_hours"], df_a2c_rewards["Value"], label="A2C", marker="o")


plt.xlabel("Training time [hours]")
plt.ylabel("Average reward")
plt.title("Training average reward - DQN vs A2C")
plt.legend()


plt.grid()
plt.show()
```
